### ***1. List Comprehension (LC)***


---



1.List Comprehension provides a shorter and cleaner way to create a new list from
an existing list.

2.It reduces the use of loops and append().

3.It improves code readability and performance.

# Syntax:
new_list = [expression for item in iterable if condition]

Common uses:

Filtering elements

Applying operations (square, uppercase, etc.)

Creating modified lists

# ***2. Lambda Function***


---



1.Lambda is an anonymous (nameless) function.

2.It can take multiple arguments.

3.It contains only one expression.

4.It automatically returns the result.

# Syntax:
lambda arguments : expression

Mainly used for short operations and with map(), filter().

# ***3. Expression in Python***
