```python
#print 1 to 10 numbers using while loop
i = 1
while i <=10:
  print(i, end = " ")
  i = i +1
```

```python
# print 1 to 10 in reverse form

i = 10
while i <=1:
  print(i, end =" ")
  i = i-1
```

```python
# find the largest number in the list -- L = [10,21,34,56,22,15,16,24]
L = [10,21,34,56,22,15,16,24]
print(max(L))
```

```python
# find the largest number in the list -- L = [10,21,34,56,22,15,16,24] 2nd method
L = [10,21,34,56,22,15,16,24]
largest = L[0]
for num in L:
    if num > largest:
        largest = num

print(largest)
```
