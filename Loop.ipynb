# Control Structure Of loop's

*   Break :- Stop the iteration's irrespective to the Next item in the collection
*   Continue:- Skip the current itteration and moved to the next
*   Pass:- is null place holder when executes nothing will happen's


```python
l = [2,4,5,6,7,8,9]
#print the all numbers expect the 5-7-9
for i in l :
  if i == 2:
    continue
  if i == 4:
    continue
  if i == 6:
    continue
  if i == 8:
    continue
  print(i)
```

```python
for i in l:
  if i == 2 or i ==4 or i ==6 or i == 8:
    continue
    # we cant stop loops at multiple data point because break statement stops the ittretion
    # can skip ittretion at multiple data points
  print(i)
```
