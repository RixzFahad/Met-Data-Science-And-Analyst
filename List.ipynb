# List Is Mutable


A collection is mutable if it follow these three Rule
**bold text**

*   We can replace or swap the item's
*   We Can Add New Item's
*   We CAn Remove The Item's






```python
# In List we can replace add , or remove the item's
```

#Repalce the Item From the List

```python
A = "Konnichiwa Minasan"
b = A.replace("Konnichiwa Minasan", "Konnbanwa")
print(b)
```

```python
List = ['Rixz', 'Roxx', 'Rix', 'Rex', 'Rax', 'Rxx']
List[3] = "Candy"
print(List)
```

#Adding the Item In the List


```python
# Append():- add the item at the last of the list
List = ['Rixz', 'Roxx', 'Rix', 'Rex', 'Rax', 'Rxx']
List.append("Candy")
print(List)
print(dir(List))
```

```python
#Inser():- Add the item at the specific location
List = ['Rixz', 'Roxx', 'Rix', 'Rex', 'Rax', 'Rxx']
List.insert(0, "Candy")
print(List)
```

```python
# Extend():- Add multiple item from another List collection
List2= ['Rixz', 'Roxx', 'Rix', 'Rex', 'Rax', 'Rxx', "Candy", "ButterFlies", ]
List2.extend(List)
print(List2)
```

# Remove Item From The List

```python
# Pop():- Here We Can Use Both Postive Or Negartive Indexing

List = ['Rixz', 'Roxx', 'Rix', 'Rex', 'Rax', 'Rxx']
List.pop(-4)
print(List)
```

```python
# Remove:- We can directly remove the item
List = ['Rixz', 'Roxx', 'Rix', 'Rex', 'Rax', 'Rxx']
List.remove("Rxx")
print(List)
```

```python
# Removin the range of item from 10, to 30.
List = ['Rixz', 'Roxx', 'Rix', 'Rex', 'Rax', 'Rxx', 10, 20, 30]
del List[5:]
print(List)
```
