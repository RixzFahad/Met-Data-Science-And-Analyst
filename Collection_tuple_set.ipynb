***Set's*** :-


*   *What are set's and how can we declare a set*
*   *The set's are declare in curly braces {}*
*   Set's are defind as unordered, immutable, and allow's the duplicates value's.

```python
#Set Allow's The Dupliacte's Data.

s = {1,2,3,4,5,6,7,7}
print(s)
# it does'nt allow the duplucates data only give one time.output for same item.

print(len(s))
# it's Also count values once if duplicates availble it dont count twice.

print(s[0])

# set's are unordered so we can't acces the values by indexing.
```

```python
# Set's Are Immutable :- In Set's We cant replace swap but we can  add or remove the item's
s = {1,2,3,4,5,6,7,7}
print(dir(s))
```

***Adding the elements in the set's***

```python
# Using the Add() Method
s = {"Rixz", "Rixx", "Rexx"}
s.add("Roxx")
print(s)
```

```python
# Using Update() Method
# This is used to add item's from another collection's

ex = {1,2,3,4}
ex.update(s)
print(ex)
```

***Remove the ITem's From The Set's***

```python
# Using POP()

Ex1 = {1, 2, 3, 4, 'Rixz', 'Rixx', 'Roxx', 'Rexx'}
Ex1.pop()
print(Ex1)
# Set's Are Unordered so dont use pop() primarily
```

```python
# Using remove()
# It's Take the item NAme to Remove
Ex1 = {1, 2, 3, 4, 'Rixz', 'Rixx', 'Roxx', 'Rexx'}
Ex1.remove(4)
print(Ex1)
```

```python
# Using Discard()
# It Also TAke the ITem_name

Rx = {1, 2, 3, 'Rixz', 'Rixx', 'Roxx', 'Rexx'}
Rx.discard("Rexx")
print(Rx)
```

***What Are The Diffrence Between Remove And Discard?***

```python
Rx.remove("deathByChocolate")
print(Rx)
# IF the item doesn't exist its throught the error
```

```python
Rx.discard("deathByChocolate")
print(Rx)
# If the item present do the querry if not go to the next querry
```

```python
#Using Clear()
Rx.clear()
print(Rx)
```

```python
# Using Del To Delete Entire Existence of a set.
del Rx
print(Rx)
```
