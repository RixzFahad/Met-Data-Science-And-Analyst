***Combining Or Mergining the DataFrame's***

```python
import pandas as pd

d1 = pd.DataFrame({
    "key": ['a', 'b', 'c', 'c', 'd', 'e'],
    "num1": range(6)
})

d2 = pd.DataFrame({
    "key": ['b', 'c', 'e', 'f'],
    "num2": range(4)
})

print(d1)
print(d2)

```

```python
print(d1)
print(d2)
```

```python
# Merge D1 And D2
pd.merge(d1,d2) #Inner Join Get the Common Data From the Both Tables
```

```python
#Get All The Information From The First Table And Matching Info From The Second Table. --Left Join
pd.merge(d1,d2,how="left")

```

```python
#Get All The Information From The First Second And Matching Info From The First Table. --Right Join
pd.merge(d1,d2,how="right")

```

```python
#Get All The Information From The Both Table.
pd.merge(d1,d2,how="outer")
```

***PIVOTING***

*Piovot table in pandas is a tool to summmerize and analyze the data.*

```python
data = [
    {"student": "Rixz", "subject": "Maths", "marks": 85, "sem": 1},
    {"student": "Rixz", "subject": "Physics", "marks": 78, "sem": 1},
    {"student": "Rixz", "subject": "Maths", "marks": 88, "sem": 2},
    {"student": "Rixz", "subject": "Physics", "marks": 82, "sem": 2},

    {"student": "Alex", "subject": "Maths", "marks": 90, "sem": 1},
    {"student": "Alex", "subject": "Physics", "marks": 86, "sem": 1},
    {"student": "Alex", "subject": "Maths", "marks": 92, "sem": 2},
    {"student": "Alex", "subject": "Physics", "marks": 89, "sem": 2},
]
```

```python
#convert above data frame into df
df = pd.DataFrame(data)
df
```

```python
#syntax: pd.pivot_table(df,values=index=,columns=,aggfunc=)
```

```python
pd.pivot_table(df,values="marks",index="student",columns="subject",aggfunc="min")
#
```

***Group By Function :-***

```python
#Group by function is applicable with aggregate function.
#Print the minimum marks for the students
df.groupby("student")["marks"].min()
```

***Applying function Df***

```python
#Appply a condition on the df those who recieve marks greater 75.
def eligibility(num):
  if num > 85:
    return "Eligible"
  else:
    return "Not Eligible"

#pd.apply()
df["marks"].apply(eligibility)
```

```python
df['eligibility'] = df['marks'].apply(eligibility)
```
