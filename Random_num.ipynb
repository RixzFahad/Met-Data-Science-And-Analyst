```python
# ================================
# NUMPY RANDOM – COMPLETE GUIDE
# ================================

import numpy as np
import matplotlib.pyplot as plt

# --------------------------------
# 1. SET RANDOM SEED (REPRODUCIBILITY)
# --------------------------------
# Using the same seed gives the same random output every time
np.random.seed(42)

# --------------------------------
# 2. UNIFORM DISTRIBUTION
# --------------------------------
# Generates random numbers uniformly in the range [0, 1)
# Shape: 3 rows and 2 columns
uniform_array = np.random.random((3, 2))

print("Uniform Distribution (3x2):")
print(uniform_array)
print()

# --------------------------------
# 3. NORMAL (GAUSSIAN) DISTRIBUTION – 1D
# --------------------------------
# Mean = 0, Standard Deviation = 1
# Generates 1000 normally distributed values
normal_1d = np.random.randn(1000)

print("Normal Distribution (1D):")
print("Mean:", normal_1d.mean())     # Should be close to 0
print("Std Dev:", normal_1d.std())   # Should be close to 1
print()

# --------------------------------
# 4. NORMAL DISTRIBUTION – 2D
# --------------------------------
# 100 rows (samples), 3 columns (features)
normal_2d = np.random.randn(100, 3)

print("Normal Distribution (100x3):")
print(normal_2d[:5])  # Print only first 5 rows
print()

# Column-wise statistics
print("Column-wise Mean:", normal_2d.mean(axis=0))
print("Column-wise Std Dev:", normal_2d.std(axis=0))
print()

# --------------------------------
# 5. RANDOM INTEGERS
# --------------------------------
# Generate a single random integer
# Range: 1 to 8 (9 is excluded)
single_int = np.random.randint(1, 9)

print("Single Random Integer:", single_int)
print()

# Generate a 3x3 matrix of random integers
int_matrix = np.random.randint(1, 9, (3, 3))

print("Random Integer Matrix (3x3):")
print(int_matrix)
print()

# --------------------------------
# 6. DIFFERENCE BETWEEN rand() AND random()
# --------------------------------
# rand() takes dimensions as separate arguments
rand_example = np.random.rand(2, 3)

# random() takes shape as a tuple
random_example = np.random.random((2, 3))

print("rand() Output:")
print(rand_example)
print()

print("random() Output:")
print(random_example)
print()

# --------------------------------
# 7. VISUALIZE NORMAL DISTRIBUTION
# --------------------------------
# Histogram helps verify bell-shaped curve
plt.hist(normal_1d, bins=30)
plt.title("Histogram of Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```
