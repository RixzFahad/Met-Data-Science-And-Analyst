***Dictinary Is Mutable***

```python
"""-> we can replace the data in dict

-> we can add new data in dict

-> we can remove the data in dict """
```

```python
#Replace the values from the dictionary

d1 = {
    "Name": "Rixz",
    "Course":"Ds Da Data Science",
    "Skills": ["Python", "Bi", "Excel"],
    "Duration": "4-6 Months"
}
#Replace the course:-
d1["Course"] = ["Web Dev"]
d1["Course"]
print(d1)
```

```python
# Adding New KEy value PAir
d1["Address"]=  "Jaganpur"
print(d1)
#if the key exist the values got replaced
#if key not exist the key get added.
```

```python
# Remove the item from the dictionary
d1 = {
    "Name": "Rixz",
    "Course":"Ds Da Data Science",
    "Skills": ["Python", "Bi", "Excel"],
    "Duration": "4-6 Months"
}
#Using POp()
d1.pop("Duration")
print(d1)
```

```python
#popitem()
d1.popitem()
print(d1)
```

***Renvision***

list --> [] --> ordered, mutable, and allows the duplicates
1. adding methods --> append, insert, extend
2. replace the items --> list[index]
3. remove methods --> pop, remove, clear, del


Tuple:- () --> ordered, immutable and allows duplicates
1. in tuple we cannot replace, add and remove the items bcz it is immutable


SET:- {}, set() --> unordered, immutable, doesn't allow duplicates

1. set is immutable we cannot replace the data but we can add and remove the items from set
2. adding methods of set --> add, update
3. remove methods --> remove, pop

Dictionary:- {key:value} --> unordered, mutable, and wont allow duplicate values

1. replace method of dictionary --> dict['existing_key_name']
2. add methods of dictionary --> dict['New_key_name'], update
3. remove methods of dictionary --> pop("key"), popitems(), clear()
