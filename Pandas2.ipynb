```python
# Difference between loc and iloc

# df.loc is label-based and supports conditional filtering
# df.iloc is integer position-based and does NOT support conditional filtering

df['col'].loc[start:end:step]

# apply condition
df.loc[df['col'] % 2 == 0]
```
