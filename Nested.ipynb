```python
# Nested Condition : A Nested Codition In Python Are the Conditios Which are inside the anothet else  or if clause.
# useful in scenarios where multiple criteria need to be checked before taking an action.

rep = 1
target = 5

if rep <= target:
    print("Rep", rep, "First Pushups Done")
    rep += 1

    if rep <= target:
        print("Rep", rep, "Second done")
        rep += 1

        if rep <= target:
            print("Rep", rep, "Third Is Done")
            rep += 1

            if rep <= target:
                print("Rep", rep, "Four Is Done")
                rep += 1

                if rep <= target:
                    print("Rep", rep, "Fifth Is Done")
```
