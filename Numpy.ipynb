# ***What Are The Diffrences Between Arrays And List.?***

```python
# 1. Data Type
# List  : Can store elements of different data types
# Array : Stores elements of the same data type

# 2. Performance
# List  : Slower for numerical computations
# Array : Faster and optimized for numerical operations

# 3. Memory Usage
# List  : Uses more memory
# Array : Uses less memory due to fixed data type

# 4. Mathematical Operations
# List  : Does not support element-wise operations directly
# Array : Supports element-wise operations

# 5. Multidimensional Support
# List  : Uses nested lists (less efficient)
# Array : Built-in support for multidimensional data

# 6. Functionality
# List  : General-purpose data structure
# Array : Designed for scientific and mathematical computing

# 7. Speed of Computation
# List  : Slower due to Python loops
# Array : Faster due to vectorized operations

# 8. Library Requirement
# List  : Built-in, no import required
# Array : Requires NumPy library
```

```python
# What is the difference between arrays and lists?

import numpy as np

# List example
l = [1, 2, 3]

# Add element to list
l.append(4)
print(l)
# Output: [1, 2, 3, 4]

# NumPy array example
e = np.array([1, 2, 3])

# Arrays do NOT support append like lists
# e.append(4)  ❌ (will cause an error)

# Correct way: element-wise operation or create a new array
f = e + np.array([4, 4, 4])
print(f)
# Output: [5 6 7]

# Key difference shown:
# List -> append adds a new element
# Array -> operations are element-wise, size is fixed
```

```python
"""
When we add a single element (scalar) to an existing NumPy array,
the scalar value is automatically applied to each element of the array.
This concept is known as the Broadcasting method.
"""

# Using Python list
l = [1, 2, 3, 4, 5, 6]

# Print list elements multiplied by 2
for i in l:
    print(i * 2, end=" ")

# Output: 2 4 6 8 10 12


# Using NumPy array (Broadcasting)
import numpy as np

g = np.array([1, 2, 3, 4, 5, 6])
print(g * 2)

# Output: [ 2  4  6  8 10 12 ]
```

# ***Define Dot Product***

```python
# Dot product is the multiplication of corresponding elements of two arrays
# followed by the sum of those products.

import numpy as np

a = np.array([1, 2])
b = np.array([3, 4])

# Easy method using NumPy
dot = np.dot(a, b)
print(dot)
# Output: 11

# Manual method
c = a * b          # element-wise multiplication
print(c)           # [3 8]
print(c.sum())     # 11
```

```python
# Dot product using different NumPy methods

import numpy as np

a = np.array([1, 2])
b = np.array([3, 4])

# Using dot() method
e = a.dot(b)
print(e)
# Output: 11

# Using @ operator (matrix / dot multiplication)
f = a @ b
print(f)
# Output: 11
```
