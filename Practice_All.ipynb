# 1Ô∏è‚É£2Ô∏è‚É£ Functions (args & kwargs)


```python
"""What Is Output For This"""
def add(*args):
    return sum(args)

print(add(1, 2, 3))

"""*args Definition"""
def add(*args):
    return sum(args)

print(add(1, 2, 3))  # 6

"""What Is Output For This"""
def info(**kwargs):
    print(kwargs)

info(name="Rixz", age=21)
```

# 1Ô∏è‚É£3Ô∏è‚É£ List Comprehension, Lambda, Map, Filter


```python
"""Convert this into list comprehension:"""
lst = []
for i in range(5):
    lst.append(i*i)
print(lst)

"""What Is Output For This"""
nums = [1, 2, 3, 4]
result = list(filter(lambda x: x % 2 == 0, nums))
print(result)
```
