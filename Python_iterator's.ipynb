# ***Iterator's In Python***

***What Are Iterator's-?***

Iterator's break the large amount of data into chunks.



***How the iterator's Can Be Build?***



Iterators can be build based of two protocol's


* 1 __iter__()
* 2 __next__()

```python

for i in range(10):
    print(i)
```

```python

l = ["Rixz", "Rox", "Rex", "Candy"]
for i in l:
    print(i)
```

```python

lx = ["Rixz", "Rox", "Rex", "Candy"]

it = iter(lx)

print(next(it))  # Rixz
print(next(it))  # Rox
print(next(it))  # Rex
print(next(it))  # Candy
```

***What Are Generator's***

* Return the values one at a time and saves the memory.
* When We Create A Generator We USe "Yield" instead of return

```python
def f1():
    yield 1
    yield 2
    yield 3
    yield 4
print(f1)
gen = f1()
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
```
