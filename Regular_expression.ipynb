# ***REGULAR EXPRESSION {RegEx}***

**Regular Expression :-**

RegEx Is  A Sequence of charcter that define search pattern
for string searching ,

 matching validation and manipulation.

```python
import re

#Library ---# Package/Module ---# Files
```

***re.search()***  :-Used when you want to:

*   Check whether a pattern exists anywhere in a string.
*   Find the first occurrence of a pattern.
*   Validate partial text.
*   Extract information from unstructured text.

***re.search(pattern, text)***



ðŸ“Œ It returns only the first match, not all matches.

```python

Text = ("Hello Candy")
Result = re.search("Candy" , Text)
print(Result)
if "Candy" in Text:
    print(Text, "Is Available")
else:
    print("Not In Your Life")
```

***re.findall()*** :-  is used to:

*   Extract multiple values from text
*   Return results as a list
*  Find ALL occurrences of a pattern in a string
*  Count how many times a pattern appears

***re.findall(pattern, string, flags=0)***



ðŸ“Œ Unlike re.search(), it does not stop at the first match.

```python

# Finding How MAny TIme Candy Repeated
Rx = ("I Know Nothing About Candy, But I LIke CAndy, Coz Candy Likes me")
Cx = re.findall("Candy", Rx)
print(Cx)
```

***re.sub() :-***  is used to:

*   Replace a pattern with another value
*   Modify text based on a regex rule

*   Clean or normalize data
*   Mask sensitive information

***re.sub(pattern, replacement, string, count=0, flags=0)***


ðŸ“Œ It replaces all occurrences by default.

```python
text3 = "Medha Is a Famous Actress"
#REplace The medha To MEdha shankar
result1 = re.sub("Medha","Medha_Shankar", text3)
print(result1)
```

***re.split() :-***  is used to:

*   Split a string based on a regex pattern
*   Break text into multiple parts

*   Perform advanced splitting (not possible with str.split)
*   Clean and tokenize data

***re.split(pattern, string, maxsplit=0, flags=0)***



ðŸ“Œ It returns a list.

```python
#Split the Text At Each And Every Space
Cx = ("Hello guys that's another day of my coding")
Cn = re.split(" ",Cx)
print(Cn)
```

*Regular Expressions are used for searching,*

 *matching, validating, extracting,*

*and manipulating text using pattern-based rules.*

```python
# Extract The Number From Above List

text4 = "The cost is $999"

x = re.search(r"\$(\d+)", text4)
print(x.group(1))
```

```python
#Extract the emails from the above list and print in a list
text5 = "Email: Rixz@gmail.com, Rox@gmail.com"
emails = re.findall(r"[a-zA-Z0-9_.]+@[a-zA-Z0-9-]+\.[a-zA-Z]+", text5)
print(emails)
```

```python

#VALIDATE A EMAIL:-
pattern = r'^\w+@\w+\.\w+$'
email = "rix@gmail.com"

if re.fullmatch(pattern, email):
    print("yes valid email")
else:
    print("invalid email")
```
